## Train with all the features

```python
train_df=pd.read_csv('train.csv', index_col=0)
test_df=pd.read_csv('test.csv', index_col=0)
```

```python
# deal with label(SalePrice), making it to be normal
y_train=np.log1p(train_df.pop('SalePrice'))
sns.distplot(y_train)
```

```python
all_df=pd.concat([train_df, test_df])
all_df.head()
```

```python
all_df['MSSubClass'].value_counts()
```

```python
# one-hot encoding
all_dummy_df=pd.get_dummies(all_df)
all_dummy_df.head()
```

```python
# Deal with missing value
all_dummy_df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
# Fill in the missing value with the mean value
mean_cols=all_dummy_df.mean()
mean_cols.head()
```

```python
all_dummy_df=all_dummy_df.fillna(mean_cols)
all_dummy_df.isnull().sum().sum()
```

```python
# Standarlize int values by using the fomula (X-X')/S
numeric_cols= all_df.columns[all_df.dtypes != 'object']
numeric_cols
```

```python
numeric_cols_mean=all_dummy_df.loc[:, numeric_cols].mean()
numeric_cols_std=all_dummy_df.loc[:, numeric_cols].std()
all_dummy_df.loc[:, numeric_cols] = (all_dummy_df.loc[:, numeric_cols]-numeric_cols_mean)/numeric_cols_std
```

```python
all_dummy_df.head()
```

```python
dummy_train_df=all_dummy_df.loc[train_df.index]
dummy_test_df=all_dummy_df.loc[test_df.index]
```

```python
dummy_train_df.shape, dummy_test_df.shape
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
```

```python
alphas=np.logspace(-3, 2, 50)
alphas
```

```python
X_train=dummy_train_df.values
X_test=dummy_test_df.values
test_scores=[]
for i in alphas:
    clf=Ridge(i)
    test_score= np.sqrt(-cross_val_score(clf, X_train, y_train, cv=10, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
    
```

```python
plt.plot(alphas, test_scores)
plt.title('alphas vs cv error')
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
N_estimators=[20, 50, 100,150, 200, 250, 300]
test_scores=[]
for i in N_estimators:
    clf=RandomForestRegressor(n_estimators=i, max_features=0.3)
    test_score= np.sqrt(-cross_val_score(clf, X_train, y_train, cv=5, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
plt.plot(N_estimators, test_scores)
plt.title('N_estimators vs CV error')
```

```python
ridge=Ridge(alpha=15)
random_forest=RandomForestRegressor(n_estimators=350, max_features=0.3)
ridge.fit(X_train, y_train)
random_forest.fit(X_train, y_train)
```

```python
ridge_predict=ridge.predict(X_test)
random_forest_predict=random_forest.predict(X_test)
ridge_score=ridge.score(X_train, y_train)
random_forest_score=random_forest.score(X_train, y_train)
print(ridge_score)
print(random_forest_score)
```

```python
y_ridge=np.expm1(ridge_predict)
y_random_forest=np.expm1(random_forest_predict)

y_final=(y_ridge+y_random_forest)/2
```

## Final Results

```python
final_df=pd.DataFrame(data={'Id': test_df.index, 'SalePrice': y_final})
```

```python
final_df.to_csv('Kaggle Competition Results.csv', index=False)
```
